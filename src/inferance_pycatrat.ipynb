```python
# 필요한 라이브러리 및 데이터 불러오기
import pandas as pd
from pycaret.classification import *
```

```python
df_train = pd.read_csv("../data/train_preprocessed.csv")
df_test = pd.read_csv("../data/test_preprocessed.csv")
```

```python
# PyCaret 설정 및 데이터 로드
clf1 = setup(data=df_train, target='대출등급')
```

```python
# 모델 비교 및 선택
best_model = compare_models()
```

```python
best_model
```

```python
# 모델 학습
final_model = create_model(best_model)
```

```python
# 모델 튜닝
tuned_model = tune_model(final_model)
```

```python
# 모델 평가
evaluate_model(tuned_model)
```

```python
# 모델 예측
predictions = predict_model(tuned_model, data=df_test)

# 마지막 결과 확인
print(predictions.head())
```
